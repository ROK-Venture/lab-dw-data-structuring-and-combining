```python
# Your code goes here
import pandas as pd
```

```python
url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'

df_insurance_1 = pd.read_csv(url_1)
df_insurance_1
```

```python
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'

df_insurance_2 = pd.read_csv(url_2)
df_insurance_2
```

```python
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

df_insurance_3 = pd.read_csv(url_3)
df_insurance_3
```

```python
df_insurance_1.columns = [col.lower().strip() for col in df_insurance_1.columns]
```

```python
df_insurance_2.columns = [col.lower().strip() for col in df_insurance_2.columns]
```

```python
df_insurance_3.columns = [col.lower().strip() for col in df_insurance_3.columns]
```

```python
df_insurance_1.rename(columns = {"st": "state"}, inplace = True)
df_insurance_1.head()
```

```python
df_insurance_2.rename(columns = {"st": "state"}, inplace = True)
df_insurance_2.head()
```

```python
df_insurance_3.head()
```

```python
df_insurance_merge = pd.concat([df_insurance_1, df_insurance_2, df_insurance_3], axis = 0, join = "outer").reset_index(drop = True)

df_insurance_merge
```

```python
df_insurance_merge.isnull().sum()
```

```python
df_insurance_clean_null = df_insurance_merge.dropna(how="all", subset =["customer"])

df_insurance_clean_null.isnull().sum()
```

```python
df_insurance_clean_null["customer"].duplicated().sum()
```

```python
df_insurance_clean_duplicat = df_insurance_clean_null.drop_duplicates(subset = ["customer"]).reset_index(drop = True)

df_insurance_clean_duplicat
```

```python
df_insurance_clean_duplicat.isnull().sum() # dont go for null values in gender beacause there are the only missing value in an almost complete datapoint/row
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df_sales = df_insurance_marketing.pivot_table(index='sales_channel', values=['total_claim_amount'], aggfunc='sum').round(2)

pivot_df_sales
```

# <span style="color:green">Insight by Robert: The agent sales channel generates the most sales. In terms of sales, this sales channel is more than twice as strong as the poor sales channel Web. However, this does not indicate which channel is the best in terms of costs. Only measured in terms of absolute turnover</span>

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_df_ltv = df_insurance_marketing.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='mean').round(2)

pivot_df_ltv
```

# <span style="color:green">Insight by Robert: The insurance company has the best customer lifetime value with people who have a high school diploma or less. These customers generally seem to be less likely to change insurance and sales should generally focus on this customer group. However, it remains to be clarified whether this group is also the most profitable group, as no claims caused have been taken into account here, which may have to be considered additionally</span>
